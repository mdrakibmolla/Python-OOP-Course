```python
# using OOP
class Robot:
    def __init__(self, robot_number):
        self.robot_number = robot_number

    def movement(self):
        print(f"Its robot {self.robot_number}")

    def fire_detect(self):
        print(f"Its robot {self.robot_number}")

    def wheel_control(self):
        print(f"Its robot {self.robot_number}")

```

```python
robot1 = Robot(robot_number=1)
```

```python
robot1.movement()
```

```python
robot1.fire_detect()
```

```python
robot1.wheel_control()
```
