## getter & setter methods in OOP

```python
class Atm:
    # constructor (special function) -> super power
    def __init__(self):
        self.pin = ""
        self.__balance = 0
        #self.menu()


    # getter method
    def get_balance(self):
      return self.__balance

    # setter method
    def set_balance(self, new_balance):
      if type(new_balance) == int:
        self.__balance = new_balance
      else:
        print("dekh bhai mazak na kar!")




    def menu(self):
        user_input = input("""
Hi How Can I Help You?
1. Press 1 to Create pin
2. Press 2 to Change Pin
3. Press 3 to Check Balance
4. Press 4 to Withdraw
5. Anything to exit
""")

        if user_input == "1":
            self.create_pin()

        elif user_input == "2":
            self.change_pin()

        elif user_input == "3":
            self.check_balance()

        elif user_input == "4":
            self.withdraw_balance()

        else:
            print("Thank you!")
            exit()

    def create_pin(self):
        user_pin = input("Enter Your Pin: ")
        self.pin = user_pin

        user_balance = int(input("Enter Balance: "))
        self.__balance = user_balance

        print("Pin Created Successfully!")
        #self.menu()

    def change_pin(self):
        old_pin = input("Enter Your Old Pin: ")

        if old_pin == self.pin:
            new_pin = input("Enter New Pin: ")
            self.pin = new_pin
            print("Pin Changed Successfully!")
        else:
            print("Incorrect Pin!")

        #self.menu()

    def check_balance(self):
        user_pin = input("Enter Your Pin: ")

        if user_pin == self.pin:
            print(f"Your Balance is {self.__balance}")
        else:
            print("Incorrect Pin!")

        #self.menu()

    def withdraw_balance(self):
        user_pin = input("Enter Your Pin: ")

        if user_pin == self.pin:
            amount = int(input("Enter Amount: "))

            if amount <= self.__balance:
                self.__balance -= amount
                print(f"You Withdraw {amount}. New Balance: {self.__balance}")
            else:
                print("Insufficient Balance!")
        else:
            print("Incorrect Pin!")

        #self.menu()
```

## Static Variable & Methods in OOP

```python
class Atm:

    # 🔥 Static Variable (Class Variable)
    __customer_id = 0

    # constructor (special function) -> super power
    def __init__(self):
        self.pin = ""
        self.__balance = 0   # 🔐 private

        # unique id assign
        self.__customer_id = Atm.__customer_id
        Atm.__customer_id += 1
        print(Atm.__customer_id)
        #self.menu()

    @staticmethod
    def get_customer_id():
      return Atm.__customer_id



    def menu(self):
        while True:
            user_input = input("""
Hi How Can I Help You?
1. Create PIN
2. Change PIN
3. Check Balance
4. Withdraw
5. Deposit
6. Exit
""")

            if user_input == "1":
                self.create_pin()

            elif user_input == "2":
                self.change_pin()

            elif user_input == "3":
                self.check_balance()

            elif user_input == "4":
                self.withdraw_balance()

            elif user_input == "5":
                self.deposit()

            else:
                print("Thank you!")
                break

    def create_pin(self):
        self.pin = input("Enter Your PIN: ")
        self.__balance = int(input("Enter Balance: "))
        print("PIN Created Successfully!")

    def change_pin(self):
        old_pin = input("Enter Old PIN: ")

        if old_pin == self.pin:
            self.pin = input("Enter New PIN: ")
            print("PIN Changed Successfully!")
        else:
            print("Incorrect PIN!")

    def check_balance(self):
        user_pin = input("Enter PIN: ")

        if user_pin == self.pin:
            print(f"Your Balance is {self.__balance}")
        else:
            print("Incorrect PIN!")

    def withdraw_balance(self):
        user_pin = input("Enter PIN: ")

        if user_pin == self.pin:
            amount = int(input("Enter Amount: "))

            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrawn {amount}. New Balance: {self.__balance}")
            else:
                print("Insufficient Balance!")
        else:
            print("Incorrect PIN!")

    def deposit(self):
        user_pin = input("Enter PIN: ")

        if user_pin == self.pin:
            amount = int(input("Enter Amount: "))
            self.__balance += amount
            print(f"Deposited {amount}. New Balance: {self.__balance}")
        else:
            print("Incorrect PIN!")
```

# Aggregation (Has a relationship)

```python
class Customer:

    def __init__(self, name, gender, address):
        self.name = name
        self.gender = gender
        self.address = address   # object pass হচ্ছে

    def print_info(self):
        print(f"Name: {self.name}, Gender: {self.gender}, address: {self.address.city}, {self.address.pin}, {self.address.state}")


class Address:

    def __init__(self, city, pin, state):
        self.city = city
        self.pin = pin
        self.state = state
```

```python
# ✅ object create (outside class)
adds = Address("Bengaluru", 45454545, "Karnataka")

cus = Customer("Bappy", "Male", adds)
```

```python
cus.print_info()
```


# Inheritance in OOP
## Inheritance allows us to define a class that inherits all the methods and properties from another class. Parent class is the class being inherited from, also called base class. Child class is the class that inherits from another class, also called derived class.
⚫ non private attribute

⚫ non private methods

```python
class User: # paren class
  def __init__(self):
    self.name = "bappy"
    self.gender = "male"

  def login(self):
    print("Login Done")

  def register(self):
    print("register Done")

class Student(User): # child class
  def __init__(self):
    self.rollno = 12

  def enroll(self):
    print("Enrolled the course")


class Instructor(User): # child class
  def __init__(self):
    self.idno = 120

  def create(self):
    print("Created the course")
```

```python
s = Student()
```

```python
s.enroll()
```

```python
s.login()
```
