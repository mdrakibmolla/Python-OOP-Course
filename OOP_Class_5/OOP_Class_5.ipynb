# Super Keyword in OOP

```python
class Phone:
  def __init__(self, price, brand, camera):
    print("inside phone copnstructor")
    self.price = price
    self.brand = brand
    self.camera = camera


  def get_price(self): # getter method
    return self._price

  def buy(self):
    print("Buying a phone inside SmartPhone Constructor")

class SmartPhone(Phone):
  def buy(self):
    print("Buying a Phone inside SmartPhone Constructor")

    super().buy() # this is super keyword use case


s = SmartPhone(200000, "apple", 3)
s.buy()

```

# Super Keyword in OOP use case-2

```python
class Phone:
  def __init__(self, price, brand, camera):
    print("inside phone copnstructor")
    self.price = price
    self.brand = brand
    self.camera = camera


class SmartPhone(Phone):
  def __init__(self, price, brand, camera, os, ram):
    print("inside SmartPhone Constructor")
    super().__init__(price, brand, camera)
    self.os = os
    self.ram = ram
s = SmartPhone(200000, "Samsung", 3, "android", 16)
print(s.brand)
print(s.camera)
print(s.os)
print(s.ram)
```

# 1. Single inheritance

```python
# Parent class
class Animal:
    def sound(self):
        print("Animal makes sound")

# Child class
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Object
d = Dog()

d.sound()  # Parent class method
d.bark()   # Child class method

#🧠 ব্যাখ্যা (very short)
#👉 Dog(Animal) মানে Dog class, Animal এর property/method পায়
#👉 তাই d.sound() কাজ করছে (parent থেকে এসেছে)
```

# 2. Multilevel inheritance

```python
# GrandFather
class Product:
  def review(self):
    print("Product Customer Review")


# Father
class Phone(Product):
  def __init__(self, price, brand, camera):
    print("inside Phone constructor")
    self.price = price
    self.brand = brand
    self.camera = camera

  def buy(self):
    print("Buying a Phone")



# child
class SmartPhone(Phone):
  pass

s = SmartPhone(200000, "samsung", 3)
s.buy()
s.review()
```

# 4. Multiple inheritance

```python
# Parent
class Phone(Phone):
  def __init__(self, price, brand, camera):
    print("inside Phone constructor")
    self.price = price
    self.brand = brand
    self.camera = camera

  def buy(self):
    print("Buying a Phone")



# Parent
class Product:
  def __init__(self):
    pass

  def review(self):
    print("Product customer review")



# child
class SmartPhone(Phone, Product):
  pass



s = SmartPhone(100000, "apple", 3)
s.buy()
s.review()
```

# Polymorphism in OOP
#### Polymorphism refers to having multiple forms. Polymorphism is a programming term that refers to the use of the same function name, but with different signatures.

```python
#Wrong Way
class Shape:
  def area(self, redius):
    return 3.14*redius**2



  def area(self, l, b):
     return l*b

s = Shape()
#s.area(2)
```

```python
#Right Way

class Shape:
  def area(self, redius, l,b):
    if b == None:
      print("Redius")
      return 3.14*redius**2
    else:
      print("Multiply")
      return l*b

```

```python
s = Shape()
s.area(redius=8, l=45, b=None)
```


# Abstraction in OOP
#### Abstraction is a fundamental concept in Python programming that allows us to simplify complex concepts and focus on the essential details. It involves hiding unnecessary details and exposing only the relevant information to the users.

```python
# Abstract Base Class (ABC)

from abc import ABC, abstractmethod

# Senior
# 🔵 Parent (Abstract Class)
class BankServer(ABC):

    def database(self):
        print("Connected to database")

    @abstractmethod
    def security(self):
      print("security passed")

    @abstractmethod
    def display(self):
      print("data displayed")



# junior
# 🟢 Child 1
class MobileApps(BankServer):

    def mobile_login(self):
        print("Mobile app login successful")

    def security(self):
        print("security passed")

    def display(self):
        print("data displayed")



#junior
# 🟢 Child 2
class WebSite(BankServer):

    def web_login(self):
        print("Website login successful")

    def security(self):
        print("security passed")

    def display(self):
        print("data displayed")


# Abstract method = security rule
# Mobile app → OTP security
# Web app → Password security
# মানে: parent বলে “security লাগবেই”, child নিজের মতো বানায়
```

```python
m = MobileApps()
```

```python
m.database()
m.security()
m.display()
```

```python
w = WebSite()
```

```python

w.database()
w.security()
w.display()
```
