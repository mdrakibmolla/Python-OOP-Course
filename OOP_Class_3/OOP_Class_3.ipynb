```python
class AtmMachine:
    # constructor (special function) -> super power
    def __init__(self):
      print(id(self))
      self.pin = ""
      self.balance = 0
      #self.menu()

    def menu(self):
        user_input = input("""
Hi How Can I Help You?
1. Press 1 to Create pin
2. Press 2 to Change Pin
3. Press 3 to Check Balance
4. Press 4 to Withdraw
5. Anything to exit
""")

        if user_input == "1":
            self.create_pin()

        elif user_input == "2":
            self.change_pin()

        elif user_input == "3":
            self.check_balance()

        elif user_input == "4":
            self.withdraw_balance()

        else:
            print("Thank you!")
            exit()

    def create_pin(self):
        user_pin = input("Enter Your Pin: ")
        self.pin = user_pin

        user_balance = int(input("Enter Balance: "))
        self.balance = user_balance

        print("Pin Created Successfully!")
        #self.menu()

    def change_pin(self):
        old_pin = input("Enter Your Old Pin: ")

        if old_pin == self.pin:
            new_pin = input("Enter New Pin: ")
            self.pin = new_pin
            print("Pin Changed Successfully!")
        else:
            print("Incorrect Pin!")

        #self.menu()

    def check_balance(self):
        user_pin = input("Enter Your Pin: ")

        if user_pin == self.pin:
            print(f"Your Balance is {self.balance}")
        else:
            print("Incorrect Pin!")

        #self.menu()

    def withdraw_balance(self):
        user_pin = input("Enter Your Pin: ")

        if user_pin == self.pin:
            amount = int(input("Enter Amount: "))

            if amount <= self.balance:
                self.balance -= amount
                print(f"You Withdraw {amount}. New Balance: {self.balance}")
            else:
                print("Insufficient Balance!")
        else:
            print("Incorrect Pin!")

        #self.menu()
```

```python
obj = AtmMachine()
```

```python
class AtmMachine:

    def __init__(self):
        self.pin = ""
        self.__balance = 0 ## Private Variable
        #self.menu()

    def __menu(self): ## Private Variable
        user_input = input("""
Hi How Can I Help You?
1. Press 1 to Create pin
2. Press 2 to Change Pin
3. Press 3 to Check Balance
4. Press 4 to Withdraw
5. Anything to exit
""")

        if user_input == "1":
            self.create_pin()

        elif user_input == "2":
            self.change_pin()

        elif user_input == "3":
            self.check_balance()

        elif user_input == "4":
            self.withdraw_balance()

        else:
            print("Thank you!")
            exit()

    def create_pin(self):
        user_pin = input("Enter Your Pin: ")
        self.pin = user_pin

        user_balance = int(input("Enter Balance: "))
        self.__balance = user_balance  ## Private Variable

        print("Pin Created Successfully!")
        #self.menu()

    def change_pin(self):
        old_pin = input("Enter Your Old Pin: ")

        if old_pin == self.pin:
            new_pin = input("Enter New Pin: ")
            self.pin = new_pin
            print("Pin Changed Successfully!")
        else:
            print("Incorrect Pin!")

        #self.menu()

    def check_balance(self):
        user_pin = input("Enter Your Pin: ")

        if user_pin == self.pin:
            print(f"Your Balance is {self.__balance}")  ## Private Variable
        else:
            print("Incorrect Pin!")

        #self.menu()

    def withdraw_balance(self):
        user_pin = input("Enter Your Pin: ")

        if user_pin == self.pin:
            amount = int(input("Enter Amount: "))

            if amount <= self.__balance: ## Private Variable
                self.__balance -= amount  ## Private Variable
                print(f"You Withdraw {amount}. New Balance: {self.__balance}")  ## Private Variable
            else:
                print("Insufficient Balance!")
        else:
            print("Incorrect Pin!")

        #self.menu()
```
